# **(4-b)**

```
admin_user = "Sarhan"
admin_pass = "2302"

tries = 3
attempt = 0

while attempt < tries:
    user = input("Enter username: ")
    pwd = input("Enter password: ")

    if user == admin_user and pwd == admin_pass:
        print(f"Login successful! Welcome, {user}.")
        break
    elif user == admin_user:
        print("Incorrect password. Try again.")
    else:
        print("User not found.")

    attempt += 1

if attempt == tries:
    print("Account locked.")
```

# **(5-a)**

```
pin = "2302"
bal = 100000

user_pin = input("Enter your PIN: ")

if user_pin == pin:
    amount = float(input("Enter withdrawal amount: "))

    if amount > 500 and amount <= bal:
        bal -= amount
        print("Withdrawal successful.")
        print("Remaining balance:", bal)
    elif amount <= 500:
        print("Minimum withdrawal amount is greater than 500.")
    else:
        print("Insufficient balance.")
else:
    print("Wrong PIN.")
```

# **(5-b)**

```
pin = "2302"
bal = 10000
tries = 4

while tries > 0:
    user_pin = input("Enter your PIN: ")

    if user_pin == pin:
        amt = float(input("Enter withdrawal amount: "))

        if amt > 500 and amt <= bal:
            bal -= amt
            print("Withdrawal successful.")
            print("Remaining balance:", bal)
        elif amt <= 500:
            print("Minimum withdrawal amount is greater than 500.")
        else:
            print("Insufficient balance.")
        break
    else:
        tries -= 1
        if tries > 0:
            print(f"Wrong pin. Attempts remaining: {tries}")
        else:
            print("Card blocked!")
```

# **(6-a)**

```
# 1 to 20
for num in range(1, 21):
    print(num)

# Multi Table
for n in range(1, 11):
    print(f"7 x {n} = {7*n}")

# Sum
sum_total = 0
for value in range(1, 101):
    sum_total += value
print("Sum from 1 to 100:", sum_total)

# Even numbers
for even in range(2, 51, 2):
    print(even)

# Fruits
fruit_list = ['apple', 'banana', 'cherry', 'mango', 'grape']
for pos, fruit in enumerate(fruit_list, start=1):
    print(pos, fruit)
```
